```python
import pandas as pd
```

```python
from pymorphy2 import MorphAnalyzer
```

```python
morph = MorphAnalyzer()
```

```python
df = pd.read_csv('my_recipes_data.csv')
ingredients = df['ингредиенты']
```

```python
# создание полного списка ингредиентов
ingredients = list(ingredients)
```

```python
# обработка данных ингредиентов (нижний регистр, лемматизация при возможности)
for odin in range(len(ingredients)):
    ingredients[odin] = ingredients[
        odin].replace("['",'').replace("']", '').replace('"',''
        ).replace(')','').replace('(','').strip().split("', '")

    for ing in range(len(ingredients[odin])):
        if len(ingredients[odin][ing].split()) == 1:
            ingredients[odin][ing] = morph.parse(ingredients[odin][ing].lower())

            for i in ingredients[odin][ing]:
                if 'NOUN' in i.tag and 'nomn' in i.tag:
                    ingredients[odin][ing] = i.normal_form
        else:
            ingredients[odin][ing] = ingredients[odin][ing].lower()
```

```python
# обработка данных ингредиентов, состоящих из нескольких слов
for i in ingredients:
    for e in range(len(i)):
        if type(i[e]) == list:
            i[e]= i[e][0].word
        if len(i[e].split(', ')) != 1:
            i.extend(i[e].split(', '))
            i.remove(i[e])
        elif i[e] == "и т.д" or i[e] == 'т.п':
            i.remove(i[e])
        elif len(i[e].split(' и ')) != 1:
            i.extend(i[e].split(' и '))
            i.remove(i[e])
        elif len(i[e].split(' или ')) != 1:
            i.extend(i[e].split(' или ')[0])
            i.remove(i[e])
```

```python
for i in ingredients:
    for e in range(len(i)):
        if type(i[e]) == list:
            i[e]= i[e][0].word
```

```python
for i in range(len(ingredients)):
    for e in range(len(ingredients[i])):
        for w in repet:
            if w in ingredients[i][e]:
                ingredients[i][e] = w.strip()
            if 'яйц' in ingredients[i][e]:
                ingredients[i][e] = 'яйцо'
            elif 'сгущ' in ingredients[i][e]:
                ingredients[i][e] = 'сгущенное молоко'
    ingredients[i] = list(set(ingredients[i]))
```

```python
clean_ingr = [ingr for ingr in clean_ingr  if len(ingr) > 1]
full = sorted(clean_ingr)
df_full = pd.DataFrame(full)
df_full.insert(0, "id", range(len(df_full)))
```
